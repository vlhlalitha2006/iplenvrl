```python
import pandas as pd, plotly.express as px
df = pd.read_csv('training/logs/rewards.csv')
avg = df.groupby('episode')['TOTAL'].mean().reset_index()
fig = px.line(avg, x='episode', y='TOTAL', title='Average Reward per Episode')
fig.show()
```
